```python
f_nom = 60
t = np.linspace(0, 2*np.pi, 50)

ma = 1.5

Va = ma*np.sin(t)
Vb = ma*np.sin(t + np.pi/3)
Vc = ma*np.sin(t + 2*np.pi/3)

V = np.array([Va, Vb, Vc]).transpose()

plt.plot(t, Va)
plt.plot(t, Vb)
plt.plot(t, Vc)
plt.show()
```

```python
T = 2/3*np.array([[1, -1/2, -1/2],
                  [0, np.sqrt(3)/2, -np.sqrt(3)/2],
                  [1/2, 1/2, 1/2]])

VT = (V @ T).transpose()
```

```python
V_alpha = VT[0]
V_beta = VT[1]

plt.plot(t, VT[0], label='$V_{alpha}$')
plt.plot(t, VT[1], label='$V_{beta}$')
plt.plot(t, VT[2], label='$V_{gama}$')
plt.legend()
plt.show()
```
