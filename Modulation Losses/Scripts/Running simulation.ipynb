# Trabalho 4 da disciplina de modulação
## Typhoon files
Arquivos utilizados para o PWM Modulation 2 níveis entregue ao HIL Academy da Typhoon.

```python
import typhoon.api.hil as hil
import typhoon.test.capture as cap

import matplotlib.pyplot as plt
import numpy as np

import pickle
```

Parâmetros fixados:

```python
res = 1
ind = 50e-6
m_index = 1
dc_volt = 800
```

**Captura de ondas do 2 níveis com modulação ``carrier_based``**

```python
if not hil.load_model(
        file='../Typhoon files/two_level_carrier_based Target files/two_level_carrier_based.cpd', 
        vhil_device=True
    ):
    print('You need to compile the model in Typhoon HIL software and give the right path')
```

```python
hil.set_scada_input_value('resistance', res)
hil.set_scada_input_value('inductance', ind)
hil.set_scada_input_value('modulation index', m_index)
hil.set_source_constant_value('Vs1', dc_volt)
```

```python
labels = {'i_out': 'Three Phase Inverter1.IGBT Leg1.il',
          'sw': 'Three Phase Inverter1.IGBT Leg1.PWM_Modulator_TOP_1'}
signals = list(labels.values())

duration = 36 * 1/60
```

```python
hil.start_simulation()

cap.start_capture(
    duration=duration,
    rate=100e3,
    signals=signals
)

captures = cap.get_capture_results(wait_capture=True)
hil.stop_simulation()

captures.index = captures.index / np.timedelta64(1, 's')
```

```python
captures[labels['i_out']].plot()
plt.show()
captures[labels['sw']].plot()
plt.show()
```

```python
captures.to_pickle('../output/carrier_based.pklcap')
```
