```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_nom = 60
t = np.linspace(0, 10*2*np.pi, 1024)

Va = 311*np.sin(t)
Vb = 311*np.sin(t + np.pi/3)
Vc = 311*np.sin(t + 2*np.pi/3)

plt.plot(t, Va)
plt.plot(t, Vb)
plt.plot(t, Vc)
plt.show()
```
